```python
try:
    del pyDict,outJson,df
except:
    pass
import pandas as pd
def df_build(raw_data:list, data_keys:list):
    df = pd.DataFrame(columns = data_keys)
    for review in raw_data:
        try:
            df.loc[len(df)] = [review[key] for key in data_keys]
        except KeyError:
            break
    return df
```

```python
from preprocess import main as preprocess
import requests
# pull single data
data_keys = ["recommendationid", "author", "review", "timestamp_created", "timestamp_updated", "voted_up", "votes_up", "votes_funny",
             "weighted_vote_score","comment_count", "steam_purchase", "received_for_free", "written_during_early_access"]
parameters = {"json": 1, "filter":"recent", "num_per_page":100, "language": "english", "purchase_type":"all", "day_range":365, "cursor":"*"}
game = 367520
url = f'https://store.steampowered.com/appreviews/{game}'
response = requests.get(url, params=parameters).json()
# data -> df
df = df_build(response["reviews"], data_keys)
print(df.index.size)
print(df.columns)
print(df.columns.size)
df.head(2)
```
